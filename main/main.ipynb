```python
from rewrite_best import rewrite_with_original, init_gemini
from retrieval_chroma import load_chroma_db, search_topk_by_text
from rrf import rrf_fusion
from rerank_GPU import RerankModel
from sentence_transformers import SentenceTransformer
from extract_model import Model_Extractor
from call_api import generate_final_answer
import google.generativeai as genai
```

```python

# Khởi tạo resource
db_path = "/content/drive/MyDrive/DS310/Vector Database"
load_chroma_db(db_path, collection_name="langchain")

RERANK_MODEL_PATH = "huynhdat543/VietNamese_law_rerank"

EXTRACT_MODEL_PATH = "lmka05/Model_finetune_bert"
subfolder_name = "checkpoint-6795"

reranker = RerankModel(RERANK_MODEL_PATH)

extractor = Model_Extractor(EXTRACT_MODEL_PATH, subfolder_name)

GOOGLE_API_KEY = ""
genai.configure(api_key=GOOGLE_API_KEY)
gemini_model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
# Pipeline function

def qa_pipeline_chroma(
    user_query: str,
    num_rewrites: int = 3,
    retrieval_k: int = 100,
    rrf_top_n: int = 50,
    rerank_top_k: int = 10,
):
    """
    Trả về Top-K đoạn phù hợp nhất cho query
    """

    # Query rewrite
    init_gemini("")
    queries = rewrite_with_original(user_query, num_queries=num_rewrites)

    # Retrieval cho từng query
    retrieval_lists = []

    for q in queries:
        results = search_topk_by_text(q, top_k=retrieval_k)
        retrieval_lists.append(results)

    # RRF Fusion
    topN = rrf_fusion(
        results_per_query=retrieval_lists,
        top_n=rrf_top_n
    )

    # Rerank bằng Cross-Encoder
    topK = reranker.rerank(
        query=user_query,
        candidates=topN,
        top_k=rerank_top_k
    )

    # Extract
    context = [doc['text'] for doc in topK]
    extract_info = extractor.extract([user_query],context)

    # Generate answer
    answer = generate_final_answer(gemini_model,user_query, extract_info[0])

    return {
        "query": user_query,
        "queries_rewrite": queries,
        "top_contexts": topK,
        "extract_info": extract_info,
        "answer": answer
    }
```

```python
result = qa_pipeline_chroma(
    user_query="Đối với những người nghỉ hưu trước năm 1995 có mức hưởng thấp, việc điều chỉnh bổ sung được thực hiện như thế nào?",
)
```

```python
print(result["queries_rewrite"])
for i, ctx in enumerate(result["top_contexts"], 1):
    print("-----")
    print("Rank", i)
    print("RRF:", ctx["rrf_score"])
    print("Rerank:", ctx["rerank_score"])
    print("Doc:", ctx["metadata"].get("SoHieu", ""))
    print(ctx["text"])
print('\n Thông tin cuối cùng trước khi đưa call API \n')
print('Câu hỏi :', result["query"])
print(result["extract_info"][0])
```
