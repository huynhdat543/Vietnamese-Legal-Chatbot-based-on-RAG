```python
import mteb
from sentence_transformers import SentenceTransformer
import pickle
import json

# Load model
model = mteb.get_model(
    "bqbbao6/vietnamese-legal-embedding",
    prompts={
        "query": "query: ",
        "document": "passage: "
    }
)

tasks = mteb.get_task("ZacLegalTextRetrieval")
results_1 = mteb.evaluate(model, tasks=tasks)
print(results_1)

with open("/kaggle/working/results_1.pkl", "wb") as f:
    pickle.dump(results_1, f)

with open("/kaggle/working/results_1.pkl", "rb") as f:
    data = pickle.load(f)

for task_result in data:
    print(f"Task: {task_result.task_name}")
    print(f"Main score: {task_result.main_score}")
    print(json.dumps(task_result.scores, indent=2))
```
