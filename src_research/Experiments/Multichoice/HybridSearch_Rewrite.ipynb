```python
# import gc
# import torch

df_groundtruth = pd.read_csv("/kaggle/input/datasets/qungngitravel/se365-gte-bge-ver2/tracnghiem_groundtruth/tracnghiem_groundtruth_NoDungSai.csv")

predictions = []

for _, row in tqdm(df_groundtruth.iterrows(), total=len(df_groundtruth), desc="Predicting", position=0, leave=True):

    pred = qa_pipeline(
    user_query= row["question"],
    answers = row["answers"],

    # rewrite
    use_rewrite = True,
    num_rewrites = 3,

    # retrieval
    retrieval_top_k = 100,

    # rewrite rrf
    use_rewrite_rrf = True,
    rewrite_rrf_top_n = 10,

    # rerank
    use_rerank = False,

    # dual rerank
    use_dual_rerank = False,
    rerank_top_k = 10,

    # extract
    use_extract = False,

    # print result
    print_result = False
)

    predictions.append(pred)

    # gc.collect()
    # torch.cuda.empty_cache()

df_groundtruth["pred_label"] = predictions
```

```python
output_path = "/kaggle/working/groundtruth_inference.csv"

df_groundtruth.to_csv(output_path, index=False)

print("Saved to:", output_path)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# df_groundtruth = pd.read_csv("/content/groundtruth_inference.csv")

y_true = df_groundtruth["correct_answer"]
y_pred = df_groundtruth["pred_label"]

accuracy_score = accuracy_score(y_true, y_pred)
precision_score = precision_score(y_true, y_pred, average="macro")
recall_score = recall_score(y_true, y_pred, average="macro")
f1_score = f1_score(y_true, y_pred, average="macro")


print("\ntracnghiem_groundtruth inference:")
print("Accuracy Score:", accuracy_score)
print("Precision Score:", precision_score)
print("Recall Score:", recall_score)
print("F1 Score:", f1_score)
```
