```python
from sentence_transformers import SentenceTransformer, evaluation
from datasets import load_dataset

MODEL_NAME = "bqbbao6/VN_legal_embedding_512"
DATASET_NAME = "another-symato/VMTEB-Zalo-legel-retrieval-wseg"

print("Đang tải dữ liệu...")
corpus_ds = load_dataset(DATASET_NAME, "corpus")["train"]
queries_ds = load_dataset(DATASET_NAME, "queries")["train"]
qrels_ds = load_dataset(DATASET_NAME, "data_ir")["test"]

print("Đang ánh xạ ID và Nội dung...")
corpus = {str(row['id']): row['text'] for row in corpus_ds}

queries = {str(row['query_id']): "query: " + row['question'] for row in queries_ds}

relevant_docs = {}
for row in qrels_ds:
    q_id = str(row['query_id'])
    c_id = str(row['corpus_id'])
    if q_id not in relevant_docs:
        relevant_docs[q_id] = set()
    relevant_docs[q_id].add(c_id)


print(f"Đang nạp mô hình và chạy đánh giá trên {len(queries)} câu hỏi...")
model = SentenceTransformer(MODEL_NAME)

evaluator = evaluation.InformationRetrievalEvaluator(
    queries=queries,
    corpus=corpus,
    relevant_docs=relevant_docs,
    name="Zalo-Legal-Final-Check",
    main_score_function="cosine"
)

results = evaluator(model)

print("\n" + "="*40)
print(f"KẾT QUẢ ĐÁNH GIÁ CUỐI CÙNG")
print("-" * 40)
print(f"NDCG@10: {results['Zalo-Legal-Final-Check_cosine_ndcg@10']:.4f}")
print(f"MRR@10:  {results['Zalo-Legal-Final-Check_cosine_mrr@10']:.4f}")
print("="*40)
```
